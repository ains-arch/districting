## Random Walk Example

```python
import numpy as np
import matplotlib.pyplot as plt 
T = np.array([[ 0 , 0.5, 0.5, 0], 
              [0.5,  0 , 0.5, 0],
              [1/3, 1/3,  0 , 1/3],
              [ 0 ,  0 ,  1 , 0]])

```

```python
#Set x0
x = [1,0,0,0]
plt.bar( ['A', 'B', 'C', 'D'], x)
step = 0
```

```python
# Run one step of chain
step = step + 1
x = np.matmul(x,T)
print(x) 
plt.bar( ['A', 'B', 'C', 'D'], x)
plt.title("Probability Distribution after " + str(step) + " transitions")
plt.show()
```
